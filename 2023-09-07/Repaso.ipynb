```python
#Funciones
def factorial(n):
    f = 1
    for i in range(1, n+1):
        f*=i
    return f
print(factorial(4))
```

```python
#Funciones (recursividad)
def factorial(n):
    if n>0:
        return n*factorial(n-1)
    else:
        return 1
print(factorial(4))
```

```python
#Clases
import random
class Dado(object):
    def __init__(self,numero_caras):
        self.numero_caras = numero_caras
        self.valor = 0
    def lanzar(self):
        self.valor = random.randint(1,self.numero_caras)
    def mostrar(self):
        print(self.valor)
    def __add__(self, other):
        return self.valor+other.valor

dado1 = Dado(6)
dado1.lanzar()
dado1.mostrar()

dado2 = Dado(12)
dado2.lanzar()
dado2.mostrar()

dado1+dado2
```
